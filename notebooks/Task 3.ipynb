```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# READ THE CSV BACK 
df_loaded = pd.read_csv('fintech_sentiment_analysis_results_2.csv')

print(f"✓ Loaded {len(df_loaded)} rows")
print(f"\nColumn names: {list(df_loaded.columns)}")
print(f"\nData types:\n{df_loaded.dtypes}")
print(f"\nFirst 3 rows:")
df_loaded.head(3)
```

```python
username = "postgres"
password = "123"
host = "localhost"
port = "5432"
database = "bank_reviews"

engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

print("Connected successfully!")
```

```python
banks_df = df_loaded[['bank_name', 'app_name']].drop_duplicates()

banks_df.to_sql(
    'banks',
    engine,
    if_exists='append',
    index=False
)

print("Banks inserted!")
```

```python
banks = pd.read_sql("SELECT * FROM banks", engine)

banks.head()
```

```python
reviews_df = df_loaded[[
    'bank_id',
    'review_text',
    'rating',
    'review_date',
    'sentiment_label',
    'sentiment_score',
    'identified_theme',
    'source'
]]
```

```python
reviews_df.to_sql(
    'reviews',
    engine,
    if_exists='append',
    index=False
)

print("Reviews inserted!")
```
